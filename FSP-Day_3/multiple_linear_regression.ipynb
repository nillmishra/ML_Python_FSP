```python
fig,ax = plt.subplots(nrows =  1 ,ncols=2 , figsize =(20,8))
ax[0].set_xlabel('Crime')
ax[0].set_ylabel('MEDV')
ax[0].set_title('Scatter plotting of crime with price')
ax[0].scatter(df['crim'] , df['medv'] , marker = 'o' , color = 'orange')
ax[0].axis([0,10,0,60])

ax[1].set_xlabel('Crime')
ax[1].set_ylabel('MEDV')
ax[1].set_title('Histogram potting of crime with price')
ax[1].hist(df['crim'] ,  color = 'green', bins = 100)
ax[1].axis([0,10,0,350])
```

```python
import warnings
warnings.filterwarnings('ignore')
value=['zero','non-zero']
frequencies=[df[df['zn']==0]['medv'].mean(),
            df[df['zn']!=0]['medv'].mean()]
sns.barplot(x = value,y= frequencies)
plt.xlabel('zn')
plt.ylabel('Mean Medv')
plt.title('Bar graph analysis')
plt.show()
```

```python
#indus plotting as per the medv
plt.scatter(df['indus'], df['medv'], color='red')
plt.xlabel('Industry')
plt.ylabel('Medv price')
plt.title('Industry with medv')
plt.show()
```

```python
#cahs analysis
df['chas'].value_counts()
```

```python
value=['zero','one']
frequencies = [df[df['chas']==0].shape[0],
               df[df['chas']==1].shape[0]]
fig=plt.figure(figsize=(20,8))
fig.add_subplot(1,2,1)
plt.title("pie chart plotting")
plt.pie(frequencies , labels=value , colors= ['pink','green'],
        explode =(0.1,0) , counterclock=False , autopct='%.2F%%' , shadow = True)

fig.add_subplot(1,2,2)
plt.title('Bar graph plotting')
sns.barplot(x=value,y = frequencies)
```

```python
#NOX is notrogen oxide
plt.scatter(df['nox'] , df['medv'] , color = 'black')
plt.xlabel('Nitrogen oxide')
plt.ylabel('medv')
plt.title("scattering the nox with medv")
plt.grid(ls='--')
plt.show()
```
