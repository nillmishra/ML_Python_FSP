### In timeseries analysis the rolling function is used to calculate rolling or moving-statistic such as movng average(mean) sums, standards deviations, etc. over a specified window of dataframes 

```python
df['Price'].plot(color='gray')
rolling_window_5.sum()['Price'].plot(color='blue')
```

```python
df['Price'].plot(color='gray')
rolling_window_5.min()['Price'].plot(color='blue')
```

```python
df['Price'].plot(color='gray')
rolling_window_5.max()['Price'].plot(color='blue')
```

```python
df['Price'].plot(color='gray')
rolling_window_5.std()['Price'].plot(color='blue')
```

# Rolling_windows = 10

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=10).mean().plot(color='green', label='10-period Rolling Mean')
# Add labels and title
plt.title('Price and 10-Period Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=10).max().plot(color='green', label='10-period Rolling Mean')
# Add labels and title
plt.title('Price and 10-Period Rolling max')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=10).min().plot(color='green', label='10-period Rolling Mean')
# Add labels and title
plt.title('Price and 10-Period Rolling Min')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=10).mean().plot(color='green', label='10-period Rolling Mean')
# Add labels and title
plt.title('Price and 10-Period Rolling STD')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
